```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G  = nx.Graph()
A = "A"
B = "B"
C = "C"
D = "D"
E = "E"
F = "F"

G.add_edges_from([(A,B, {'weight' : 1}), (A,C, {'weight' : 2 }), (B,C, {'weight' : 5 }), (B,D, {'weight' : 10 }), (C,D, {'weight' : 3 }), (C,E, {'weight' : 8}), (D,E, {'weight' : 4 }), (D,F, {'weight' : 6 }), (E,F, {'weight' : 1 }) ])
```

```python
#Graf Ã‡iziyorum
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels = True, font_weight = 'bold' ,  node_size=700, node_color='red', font_size= 10, font_color = 'white', font_family = 'sans-serif')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels = labels)
plt.show()
```

```python
mst = nx.minimum_spanning_tree(G)

post_mst = nx.spring_layout(mst)
nx.draw(mst, post_mst, with_labels = True, font_weight = 'bold' ,  node_size=800, node_color='blue', font_size= 10, font_color = 'black', font_family = 'Arial')

plt.show()
```

```python

post_mst = nx.spring_layout(mst)
nx.draw(mst, post_mst, with_labels = True, font_weight = 'bold' ,  node_size=800, node_color='blue', font_size= 10, font_color = 'black', font_family = 'Arial')
mst_labels = nx.get_edge_attributes(mst, 'weight')
nx.draw_networkx_edge_labels(mst, post_mst, edge_labels = mst_labels)
plt.show()
```

```python
total_weight  = sum(mst_labels.values())
print(f"Toplam AÄŸÄ±rlÄ±k: {total_weight}")
```
